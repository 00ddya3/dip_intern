```python
#중복행 제거
df2 = df2.drop_duplicates(subset=['사업명'] ,keep='first')   # 사업명을 기준으로 중복 데이터 19개 삭제

#텍스트 정제
df2['제안취지'] = df2['제안취지'].str.replace(pat=r'[^\w]',repl=r' ',regex=True)    # 특수문자 제거
df2['사업내용'] = df2['사업내용'].str.replace(pat=r'[^\w]',repl=r' ',regex=True)    # 특수문자 제거
df2['제안취지'] = df2['제안취지'].str.replace('br', '')   # html 줄내림 표시 제거
df2['사업내용'] = df2['사업내용'].str.replace('br', '')   # html 줄내림 표시 제거
df2['사업내용'] = df2['사업내용'].str.replace('ㅁ', '')   # ㅁ 제거

df2.head()
```

```python
from tqdm import tqdm
import re
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.feature_extraction.text import CountVectorizer
from konlpy.tag import Okt
from pykospacing import Spacing

okt = Okt()
spacing = Spacing()

content = df2['사업내용'].tolist()

#불용어 읽어오기
stopwords = pd.read_csv("stopwords.txt")
stopwords = stopwords['word'].tolist()

#불용어제거 + 토큰화
tks = []
for sentence in tqdm(content) :
    kospacing_sent = spacing(sentence) # 띄워쓰기 보완
    tokenized_sent = okt.morphs(kospacing_sent) # 토큰화
    stopwords_removed_sentence = [word for word in tokenized_sent if not word in stopwords] # 불용어 제거
    tks.append(stopwords_removed_sentence)

clean_sentences = []
for tk in tks :
    clean_sentences.append(" ".join(tk))

clean_sentences[:3]
```
