```python
import pandas as pd
import numpy as np
import re
import ipykernel

df1 = pd.read_csv('./rawdata/sincheon_minwon.csv')
df1.drop(columns=['Unnamed: 0'], inplace=True)
df1.head()
```

```python
df1.info()
```

```python
#중복행 제거
df1 = df1.drop_duplicates(subset=['사업명'] ,keep='first')   # 사업명을 기준으로 중복 데이터 18개 삭제

#텍스트 정제
df1['제안취지'] = df1['제안취지'].str.replace(pat=r'[^\w]', repl=r'', regex=True)    # 특수문자 제거
df1['사업내용'] = df1['사업내용'].str.replace(pat=r'[^\w]', repl=r'', regex=True)    # 특수문자 제거
df1['제안취지'] = df1['제안취지'].str.replace('br', '')   # html 줄내림 표시 제거
df1['사업내용'] = df1['사업내용'].str.replace('br', '')   # html 줄내림 표시 제거
df1['사업내용'] = df1['사업내용'].str.replace('ㅁ', '')   # ㅁ 제거

df1.head()
```

## Match Department

First by using TF-IDF, pick keyword of each department.<br>
And then, match department if no_team content has same words over 20.

```python
df1['text'] = df1['사업명'] + ' ' + df1['제안취지'] + ' ' + df1['사업내용'] + ' '

df_team = df1[df1['부서'].notnull()].reset_index(drop=True)
df_noteam = df1[df1['부서'].isnull()].reset_index(drop=True)
```

```python
df_team.head()
```

```python
# make new df with project contents according to department
L_team = list(df_team['부서'].unique())
D_team = pd.DataFrame({'부서':L_team})
D_team['민원합산'] = ''

for i in range(D_team.shape[0]) :
    text = ''
    team = D_team.loc[i, '부서']

    for j in range(df_team.shape[0]) :
        if df_team.loc[j, '부서'] == team :
            text = text + df_team.loc[j, 'text']
    
    D_team.loc[i, '민원합산'] = text

D_team.head()
```

```python
from tqdm import tqdm
import re
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt
from pykospacing import Spacing

okt = Okt()
spacing = Spacing()

#불용어 읽어오기
stopwords = pd.read_csv("stopwords.txt")
stopwords = stopwords['word'].tolist()

"""
#외래어 읽어오기
loanwords = pd.read_csv('loanwords.txt', encoding = 'cp949')
loanwords = loanwords['word'].tolist()
for word in loanwords :
    okt.add_dictionary(word, 'Noun')
"""
```

```python
# remove stopwords + tokeninzing
content = D_team['민원합산'].tolist()
D_team['민원토큰'] = ''

tks=[]
for i in range(len(content)) :
    sentence = content[i]
    kospacing_sent = spacing(sentence) # 띄워쓰기 보완
    noun_sent = okt.nouns(kospacing_sent) # 명사만 추출
    over_two_sent = [word for word in noun_sent if len(word) > 1]   # 2글자 이상인 단어만 사용
    stopwords_removed_sentence = [word for word in over_two_sent if not word in stopwords] # 불용어 제거
    D_team.at[i, '민원토큰'] = stopwords_removed_sentence
    tks.append(stopwords_removed_sentence)

clean_sents = []
for tk in tks :
    clean_sents.append(" ".join(tk))

D_team.head()
```

```python
#get main words for each department
import time

tv = TfidfVectorizer()
tv_array = tv.fit_transform(clean_sents).toarray()  #문장별 단어 tfidf 값을 가진 [46,3464] 배열
tv_vocab = {v:k for k,v in tv.vocabulary_.items()}  #단어고유번호:단어 딕셔너리

team_keywords=[]    #부서별 중요단어 리스트

for i, sent in enumerate(tv_array):
    print('====== document[%d] ======' % i)
    main_words=[]   #(단어, tfidf 지수)
    team_words=[]   #단어

    for j in range(len(sent)) :
        
        if sent[j] > 0.1 :
            ww = tv_vocab.get(j)    #keyword
            vv = sent[j]            #tfidf score
            main_words.append((ww,vv))
            team_words.append(ww)
            
    print(len(main_words), ' : ', main_words)
    team_keywords.append(team_words)
    time.sleep(1)
    
```

```python
keyword_df = D_team['부서']
keyword_df = keyword_df.to_frame()
keyword_df['주요단어'] = ''

for i in range(len(team_keywords)) :
    keyword_df.at[i, '주요단어'] = team_keywords[i]

keyword_df.head()
```
