```python
import pandas as pd
import re
import ipykernel

df1 = pd.read_csv('./rawdata/sincheon_minwon.csv')
df1.drop(columns=['Unnamed: 0'], inplace=True)
df1.head()
```

```python
#중복행 제거
df1 = df1.drop_duplicates(subset=['사업명'] ,keep='first').reset_index(drop=True)   # 사업명을 기준으로 중복 데이터 18개 삭제

#모든 내용을 담은 열 생성
df1['text'] = df1['사업명'] + ' ' + df1['제안취지'] + ' ' + df1['사업내용'] + ' '

#텍스트 정제
df1['text'] = df1['text'].apply(lambda x : re.sub(r'[^가-힣]',' ', x))  #한글 빼고 전부 제거

df1.head()
```

```python
df2 = pd.read_csv('./rawdata/국토교통부_전국 법정동_20211217.csv', encoding='euc_kr')

#대구 법정동만 가져오기
df2 = df2[df2['시도명'] == '대구광역시'][['시도명', '시군구명', '읍면동명']]
df2 = df2.fillna("")

# '@@로' 나 '@@가'로 끝나는 단어 제거
df2['읍면동명2'] = df2['읍면동명'].apply(lambda x : re.compile("[가-힣]+동").findall(x))
df2['읍면동명2'] = df2['읍면동명2'].apply(lambda x : "".join(x))

#중복값 제거
dong_list = list(set(df2['읍면동명2'].tolist()))

# 'text'열에서 @@동으로 끝나는 단어만 찾아서 리스트 타입으로 '행정구'열에 저장
df1['행정동'] = df1['text'].apply(lambda x : re.compile("[가-힣]+동").findall(x))

# 중복값 제거
df1['행정동'] = df1['행정동'].apply(lambda x : list(set(x)))

# 행정구 명이 아닌 다른 값 제거
def fdong(alldong) :
    realdong=[]

    for dong in alldong :
        if dong in dong_list :
            realdong.append(dong)

    return realdong

df1['행정동'] = df1['행정동'].apply(lambda x : " ".join(fdong(x)))

df1.head()
```

데이터가 많은 신천동, 중동, 파동, 상동, 침산동 정도만 알아보자

```python
from ckonlpy.tag import Twitter
from pykospacing import Spacing

twt = Twitter()
spacing = Spacing()

#불용어 읽어오기
stopwords = pd.read_csv("stopwords.txt")
stopwords = stopwords['word'].tolist()

#외래어 읽어오기
loanwords = pd.read_csv('loanwords.txt', encoding = 'cp949')
loanwords = loanwords['word'].tolist()
for word in loanwords :
    twt.add_dictionary(word, 'Noun')

#중립어 읽어오기
neutralwords = ['대구', '시민', '주민', '지역', '마을', '공간', '사업', '인근', '도시', '치하','천원', '억원', '만원', '여명']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from collections import Counter

dong5 = ['신천동', '파동', '중동', '상동', '침산동']

for dong in dong5 :
    
    #선택한 행정동명이 들어간 행만 추출
    df3 = df1[df1['행정동'].str.contains(dong)].reset_index(drop=True)
    content = df3['text'].tolist()

    tks=[]
    for sentence in content :
        kospacing_sent = spacing(sentence) # 띄워쓰기 보완
        noun_sent = twt.nouns(kospacing_sent) # 명사만 추출
        over_two_sent = [word for word in noun_sent if len(word) > 1]   # 2글자 이상인 단어만 사용
        stopwords_removed_sent = [word for word in over_two_sent if not word in stopwords] # 불용어 제거
        neutral_removed_sent = [word for word in stopwords_removed_sent if not word in neutralwords] # 중립어 제거
        tks.append(neutral_removed_sent)

    tk = Tokenizer()
    tk.fit_on_texts(tks)

    top15 = Counter(tk.word_counts).most_common(15)
    print("==========" + dong + "==========")
    print(top15)
```
